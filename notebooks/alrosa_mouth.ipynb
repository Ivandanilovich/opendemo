```python
import cv2
from AlrosaDemo.ImageProcessor import ImageProcessor
from AlrosaDemo.SSDDetector import SSDDetector, SSDBox
import matplotlib.pyplot as plt
ssd = SSDDetector('../models/palm_detection_builtin.tflite')
```

```python
imageProcessor = ImageProcessor()
image = plt.imread('C:/Users/ivand/Desktop/AlrosaDemo/dataset/alrosa_mouth_frames/400.jpg')
print(image.shape)
image = image[200:800, 500:1100]
# image = cv2.rotate(image, 1)

original_image, padded_image, norm_image, pad = imageProcessor.preprocess_image(image)

plt.imshow(padded_image)
plt.show()
```

```python
d = ssd.predict(norm_image)
for i in d:
    box = SSDBox(i, pad, padded_image.shape)
    image = box.vis_on_image(image)
cv2.imshow('s', image)
cv2.waitKey()
```
